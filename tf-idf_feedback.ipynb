# **TF-IDF (Term Frequency - Inverse Document Fraquency)**

*   **TF(단어 빈도, term frequency)**는 특정한 단어가 문서 내에 얼마나 자주 등장하는지를 나타내는 값. 이 값이 높을수록 문서에서 중요하다고 생각할 수 있다.
*   하지만 하나의 문서에서 많이 나오지 않고 다른 문서에서 자주 등장하면 단어의 중요도는 낮아진다.
*   **DF(문서 빈도, document frequency)**라고 하며, 이 값의 역수를 IDF(역문서 빈도, inverse document frequency)라고 한다.
*   TF-IDF는 TF와 IDF를 곱한 값으로 점수가 높은 단어일수록 다른 문서에는 많지 않고 해당 문서에서 자주 등장하는 단어를 의미한다.

```python
vect = CountVectorizer(analyzer = 'train_data', 
                             lowercase = True,
                             tokenizer = None,
                             preprocessor = None,
                             stop_words = 'english', # 불용어 제거
                             min_df = 2, # 토큰이 나타날 최소 문서 개수로 오타나 자주 나오지 않는 특수한 전문용어 제거에 좋다. 
                             ngram_range=(1, 3),
                             #vocabulary = set(words.words()), # nltk의 words를 사용하거나 문서 자체의 사전을 만들거나 선택한다. 
                             max_features = 90000
                            )
```

```python
input_data['boilerplate'] = input_data['boilerplate'].str.replace(",", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace(".", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("\"title\":", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("\"body\":", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("\"url\":", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("\"", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("1", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("2", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("3", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("4", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("5", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("6", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("7", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("8", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("9", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("0", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("\\", " ")
input_data['boilerplate'] = input_data['boilerplate'].str.replace("-", " ")
input_data['boilerplate'][0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer(stop_words="english")
tfvect = vect.fit_transform(input_data['boilerplate'])
```

# **classfication & accurancy**

```python
#  0 1 혹은 Y N Classifcation -> Binary Classifcation
from sklearn.linear_model import SGDClassifier  # logistic regression (-> Binary Classificaiton 가능)
model = SGDClassifier()
model.fit(tfvect, list(input_data['label']))  # train_X =tfvect, train_y = input_data['label']
```

```python
X_pred = vect.transform(input_data['boilerplate'])
y_pred = model.predict(X_pred)
print(y_pred)
```
